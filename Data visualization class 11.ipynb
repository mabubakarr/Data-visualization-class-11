```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://archive.ics.uci.edu/static/public/186/wine+quality.zip"
```

```python
import requests
```

```python
response = requests.get(url)
```

```python
with open("wine.zip", "wb") as f:
    f.write(response.content)
```

```python
import zipfile
```

```python
with zipfile.ZipFile("wine.zip") as zipped:
    zipped.extractall("wine")
```

```python
red_wine = pd.read_csv("wine\winequality-red.csv", sep = ";")
white_wine = pd.read_csv("wine\winequality-white.csv", sep = ";")
```

```python
red_wine.head()
```

```python
white_wine.head()
```

```python
red_corr = red_wine.corr()["quality"][:-1]
red_corr
```

```python
white_corr = white_wine.corr()["quality"][:-1]
white_corr
```
